```python
import pandas as pd
from sqlalchemy import create_engine, MetaData
import numpy as np
```

## Creating DB Engine

```python
SERVER = 'awpsg\\SQLEXPRESS'
DATABASE = 'yelpdb'
DRIVER = 'ODBC Driver 17 for SQL Server'
connecting_string = f"mssql+pyodbc://@{SERVER}/{DATABASE}?trusted_connection=yes&driver={DRIVER}"

engine = create_engine(connecting_string)
```

```python
# Create a MetaData object
metadata = MetaData()

# Reflect all tables from the database
metadata.reflect(bind=engine)

# Get the table names
table_names = metadata.tables.keys()

# Print the table names
print(table_names)
```

```python
with engine.connect() as conn:
    for each in table_names:
        query = f"SELECT TOP 5 * FROM {each}"
        df = pd.read_sql_query(query, conn)
        #display is a ipython rich output function
        display(df)
```
